一个问题的最优解可以由其子问题的最优解组合而成

```python
def fib(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fib(n-1,memo) + fib(n-2,memo)
    return memo[n]

fib(40)
```

```python
def fib(n):
    if n <= 1:
        return n
    dp = [0, 1]
    for i in range(2, n + 1):
        dp.append(dp[i - 1] + dp[i - 2])
    return dp[n]

fib(40)
```

```python
def fib(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

fib(40)
```

DP表
行数：第一个字符串的长度 +1
列数：第二个字符串的长度 +1
第一行第一列都是0
dp[i][j] 就是s1[i] 和 s2[j] 之前的最长公共子序列的长度
每个字符进行匹配，
1. 如果相等，取左上角加1
2. 不相等，取左边和上边的最大值，就是跳过这个字符

最后右下角就是最长公共子串的长度

```python
s1 = "ABCBDAB"
s2 = "BDCABA"

m, n = len(s1), len(s2)

# 初始化DP
dp = [[0] * (n + 1) for _ in range(m + 1)]

# 填表
for i in range(1, m + 1):
    for j in range(1, n + 1):
        if s1[i-1] == s2[j-1]:
            # 相同 就是左上角的值+1
            dp[i][j] = dp[i - 1][j - 1] + 1
        else:
            # 上面和左面的值，取最大
            dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

# 回溯
i, j = m, n  # 定位右下角
sub_sequence = ""

while i > 0 and j > 0:
    if s1[i - 1] == s2[j - 1]:
        sub_sequence = s1[i - 1] + sub_sequence
        # 走左上角
        i -= 1
        j -= 1
    elif dp[i - 1][j] >= dp[i][j - 1]:
        i -= 1  # 向上走
    else:
        j -= 1 # 向左走
        
print(dp[m][n], sub_sequence)
```

```python
# 打家劫舍
# n <= 2 可以直接算
# 新加的房子 偷：f(i-2) + sum[i]
# 不偷：f(i-1)
# 取最大值 赋值f(i)
def rob(houses):
    if not houses:
        return 0
    if len(houses) == 1:
        return houses[0]
    dpi_1 = max(houses[0], houses[1])
    dpi_2 = houses[0]
    for i in range(2, len(houses)):
        dpi = max(dpi_1, dpi_2 + houses[i])
        dpi_2 = dpi_1
        dpi_1 = dpi
    return dpi

rob([2, 7, 9, 3, 1, 5, 8])
```

```python
# Kadana 算法
# current_sum = max(today, current_sum+today)
# max_sum = max(max_sum, current_sum)

# 最大子数组和

def max_sub_array_sum(nums):
    curr_sum = max_sum =nums[0]
    for num in nums[1:]:
        curr_sum = max(num,curr_sum+num)
        max_sum = max(curr_sum,max_sum)
    return max_sum

max_sub_array_sum([1,2,-2,4,5])
```

```python
# 零钱兑换
# 凑大数，先试试每个面值的一个硬币，可以直接使用之前计算的最小值+1
# 当然只能选面值小于大数的硬币
# 每个面值 对应的 min(硬币数)


def exchange(coins, amount):
    if not coins:
        return -1
    if amount == 0:
        return 0
    dp = [float("inf")] * (amount + 1)
    dp[0] = 0
    for i in range(1, amount + 1):
        for coin in coins:
            if i - coin >= 0:
                dp[i] = min(dp[i],dp[i-coin]+1)
    
    if dp[amount] == float('inf'):
        return -1
    else:
        return dp[amount]


coins = [1, 2, 5]
amount = 11

exchange(coins,amount)
```

```python
# 0/1 背包问题
items_weight = [1, 3, 4, 5]
items_value = [15, 20, 30, 50]


def pack(items_weight, items_value, pack_limit):
    n = len(items_weight)
    dp = [[0] * (pack_limit + 1) for _ in n]

    # 先处理第0行，第一个物品
    for i in range(pack_limit + 1):
        if i >= items_weight[0]:
            dp[0][i] = items_value[0]
        else:
            continue

    # 填充剩余行
    for i in range(1, n):
        for j in range(pack_limit + 1): # 背包还剩多少容量
            not_pick = dp[i - 1][j]  # 不选当前物品
            pick = 0
            if items_weight[i] <= j:
                pick = dp[i - 1][j - items_weight[i]] + items_value[i]
            dp[i][j] = max(pick, not_pick)
    return dp[n - 1][pack_limit]
```

```python
# 编辑距离
# 插入，删除，替换
# 匹配不替换


def l_distance(s1, s2):
    if not s1:
        return len(s2)
    if not s2:
        return len(s1)
    # if len(s1) > len(s2):
    #     s1, s2 = s2, s1

    dp = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]
    # dp[0][0] = 0

    # 初始化边界情况
    for i in range(len(s2) + 1):
        dp[0][i] = i  # 删除操作
    for i in range(len(s1) + 1):
        dp[i][0] = i  # 插入操作

    for i in range(1, (len(s1) + 1)):
        for j in range(1, (len(s2) + 1)):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                insert_step = dp[i][j - 1]  # 从左边来
                delete_step = dp[i - 1][j]  # 从上边来
                replace_step = dp[i - 1][j - 1]  # 从左上来

                dp[i][j] = min(insert_step, delete_step, replace_step) + 1

    i, j = len(s1), len(s2)
    operation = []

    while j > 0 or i > 0:
        if i > 0 and j > 0 and s1[i - 1] == s2[j - 1]:
            operation.append(f"保留{s1[i-1]}")
            i -= 1
            j -= 1
        else:
            # 替换
            if i > 0 and j > 0 and dp[i][j] == dp[i - 1][j - 1] + 1:
                operation.append(f"{s1[i-1]}替换成{s2[j-1]}")
                i -= 1
                j -= 1
            # 删除
            if i > 0 and dp[i][j] == dp[i - 1][j] + 1:
                operation.append(f"删除{s1[i-1]}")
                i -= 1
            # Insert
            if j > 0 and dp[i][j] == dp[i][j - 1] + 1:
                operation.append(f"插入{s2[j-1]}")
                j -= 1
    operation.reverse()
    print(operation)

    return dp[len(s1)][len(s2)]


l_distance("horse", "ros")
```

```python
#LIS

def lengthOfLIS(nums):
    n = len(nums)
    if not nums:
        return 0
    #初始化dp
    dp = [1] * n
    
    for i in range(1, n):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j]+1)
    
    return max(dp)

lengthOfLIS([186,186,150,200,160,130,197,200])
```

```python
# 二分查找
import bisect
def len_LIS(nums):
    tails=[]
    
    for num in nums:
        if not tails:
            tails.append(num)
        elif tails[-1] < num:
            tails.append(num)
        else:
            pos = bisect.bisect_left(tails,num)
            tails[pos] = num
    return len(tails)

len_LIS([186,186,150,200,160,130,197,200])
```
