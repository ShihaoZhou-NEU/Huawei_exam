```python
# 矩阵乘法

a = [[1, 2, 3], [4, 5, 6]]
b = [[1, 2], [3, 4], [5, 6]]

x = len(a)
y = len(b)
z = len(b[0])

b_t = list(zip(*b))
# 结果需要初始化
c = [[] for _ in range(x)]
for i in range(x):
    for j in range(z):
        # 先要算一个单位的list，也就是一行
        sums = [n1*n2 for n1,n2 in zip(a[i],b_t[j])]
        print(sums)
        print(sum(sums))
        # 每行进行添加
        c[i].append(sum(sums))
        
c
        
```
