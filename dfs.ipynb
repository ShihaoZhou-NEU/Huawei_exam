```python
#括号生成
res = []
def dfs(paths, n,left_num,right_num):
    if left_num>n or left_num<right_num:
        return
    
    if len(paths) == n*2:
        return res.append(paths)
    
    dfs(paths+"(",n, left_num+1,right_num)
    dfs(paths+")",n, left_num,right_num+1)

def generatebucket(n):
    dfs("",n,0,0)
    print(res)

generatebucket(2)
```
